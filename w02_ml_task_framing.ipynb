## 1. My lane as an ML task (type)

*Task Type: Scoring and Ranking.
Why: We are not just classifying pages as "good" or "bad". Instead, we need to predict the "Expected CTR" for a page based on its features (search volume, intent, position tier). By subtracting the actual CTR from this predicted expected CTR, we generate an "Opportunity Score" for each page. Finally, we rank the pages based on this score to tell the content team exactly which pages to prioritize first.*

## 2. Target or proxy

*Target: The target we are predicting is the Expected CTR for a given page.
Where the label comes from: The label comes from an observed outcome. We derive it from historical data by calculating the actual CTR (clicks_90d / impressions_90d) of peer pages that share similar characteristics (such as the same position_tier, main_intent, and competition_level). The average observed CTR of these healthy peer pages serves as our proxy for what the baseline "expected" CTR should be.*

## 3. Success metric

*Success Metric: CTR Lift (or Incremental Clicks) on the intervened pages.
What means 'good': 'Good' is achieved if the top-ranked "high opportunity" pages identified by our model show a statistically significant increase in their actual CTR after the content team updates their meta tags, resulting in measurable incremental clicks over the following 30 to 90-day period.*

## 5. Why ML beats a fixed rule here

*Why ML beats a fixed rule:
A simple fixed rule (like if CTR < 10% then flag) is too rigid and ignores the complex reality of search behavior. A 5% CTR might be terrible for a low-competition transactional keyword, but that exact same 5% could be excellent for a highly competitive informational query ranking at the bottom of Page 1. The pattern is too messy for nested if-statements because variables like search_volume, competition_level, main_intent, and exact rankings interact in highly dynamic, non-linear ways. Machine Learning can simultaneously weigh all these features to calculate a customized, dynamic "Expected CTR" for every single piece of content, whereas a fixed rule would result in countless false positives and missed opportunities.*
