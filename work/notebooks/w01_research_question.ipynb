## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# Doğrudan senin GitHub repondaki dosyanın canlı linkini okutuyoruz
data_url = "https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv"

# Veri setini yüklüyoruz
df = pd.read_csv(data_url)

# Sadece yeterli gösterimi olan (visible) sayfaları filtreleyelim
visible_pages = df[df['impressions_90d'] >= 500]

# 1. sayfada (page_1) yer alıp CTR'ı %10'un (0.10) altında kalan sayfalar
page_1_low_ctr = visible_pages[(visible_pages['position_tier'] == 'page_1') & (visible_pages['ctr'] < 0.10)]

# Sayıları ekrana yazdırıyoruz
print(f"Total visible pages (impressions >= 500): {len(visible_pages)}")
print(f"High Opportunity: Pages ranking on Page 1 but with unusually low CTR (<10%): {len(page_1_low_ctr)}")
print("\nWhy this matters: These are 'low-hanging fruits'. They already have Google's trust (Page 1), but users aren't clicking. Fixing them is our capstone's goal.")
```

*What I can claim: I can identify observed directional gaps between a page's actual CTR and the expected CTR for its position tier. I can provide a data-backed, ranked list to support human decision-making.
